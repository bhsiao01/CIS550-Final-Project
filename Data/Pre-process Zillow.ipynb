# Metro Zillow

```python
zillow = pd.read_csv('Metro Zillow.csv')
```

```python
zillow
```

```python
clean_zillow = zillow.melt(id_vars=zillow.columns[0:5], 
        var_name="Date", 
        value_name="Value")
```

```python
clean_zillow
```

```python
clean_zillow.sort_values(['RegionID','Date'],inplace=True)
```

```python
clean_zillow = clean_zillow[~clean_zillow['Value'].isna()]
```

```python
clean_zillow['City'] = clean_zillow['RegionName'].apply(lambda x: x.split(',')[0])
```

```python
clean_zillow['State'] = clean_zillow['StateName']
clean_zillow
```

```python
clean_zillow['Metro'] = clean_zillow['City']
clean_zillow.drop('RegionName',axis=1,inplace=True)
clean_zillow.rename({'City':'RegionName'},axis=1,inplace=True)
```

```python
clean_zillow['CountyName'] = np.nan
```

```python
clean_zillow=clean_zillow[['RegionID','SizeRank','RegionName','RegionType','StateName','State','Metro','CountyName','Date','Value']]
```

```python
clean_zillow
```

```python
clean_zillow.to_csv('MetroZillowData.csv')
```
