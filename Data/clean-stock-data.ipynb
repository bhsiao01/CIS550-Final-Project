```python
import pandas as pd
import glob
import os

li = []

def getStocks(path):
    all_files = glob.glob(path + "/*.csv")
    
    for filename in all_files:
        df = pd.read_csv(filename, index_col=None, header=0)
        ticker_symbol = os.path.basename(filename).replace('.csv', '')
        df['StockSymbol'] = ticker_symbol
        li.append(df)
    
getStocks(r'C:\Users\Jasmine\stonks\stocks')
getStocks(r'C:\Users\Jasmine\stonks\etfs')
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame.to_csv('all_stocks.csv', index=False)
```
